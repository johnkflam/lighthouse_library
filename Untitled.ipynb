```python
grp_topics = df.groupby('topics', sort=True)
```

```python
output=open('browse_title.html','w')
def out(msg):
    output.write(msg+'\n')
    #print(msg+'\n')
```

```python
cnt=10
for k,v in grp_topics:
    id='id'+str(cnt)
    num_of_books=len(v)
    cnt += 1
    out('<div class="panel-group">')
    out('<div class="panel panel-default">')
    out('<div class="panel-heading">')
    out('<h4 class="panel-title">')
    out('<a data-toggle="collapse" href="#{0}">{1}  ({2})</a>'.format(id,k.strip(),num_of_books))
    out('</h4>')
    out('</div>')
    v.sort_values(by='title')
    out('<div id="{0}" class="panel-collapse collapse">'.format(id))
    out('<ul class="list-group">')
    for row, data in v.iterrows():
        out('<li class="list-group-item">{0}</li>'.format(data['title']))
    out('</ul>')  
    out('</div>')
    out('</div>')
    out('</div>')
```
